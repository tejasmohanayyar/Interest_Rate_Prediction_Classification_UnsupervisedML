## `Segmenting customers into various interest rate categories using USL.`

```python
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cluster_data = pd.read_csv('Data/filled_train_28-03.csv').drop(['Interest_Rate','Months_Since_Deliquency'],axis = 1)
```

```python
#Standardize the cluster data
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
data_scaled = pd.DataFrame(ss.fit_transform(cluster_data.select_dtypes(np.number)),columns = cluster_data.select_dtypes(np.number).columns)

#Perform One Hot encoding on Catefgorical columns and append with data_scaled
cat_encoded = pd.get_dummies(cluster_data.select_dtypes('object'),drop_first = True)

#Concatenating the categorical encoded columns and the scaled data
data_final = pd.concat((data_scaled,cat_encoded),axis = 1)

# #Viewing the final dataset
data_final.head()
```

```python
from sklearn.cluster import KMeans
wcv = []
for i in range(1,10):
    km = KMeans(n_clusters = i)
    km.fit(data_final)
    wcv.append(km.inertia_)

plt.figure(figsize = (15,6))
plt.plot(range(1,10),wcv,color = 'g')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
data = data_final.sample(30000)
plt.figure(figsize = (15,6))
sch.dendrogram(sch.linkage(data,method = 'ward'),p =10,truncate_mode = 'level')
plt.show()
```

### Clustering Inferences (Without Months_Since_Deliquency)

* The WCV score shows us the optimum number of clusters as 2,3 or 4. We will be going with 3 since this way we will also be able to compare the quality of the target and how well it is seperated based on this dataset.

* Due to system limitations I have taken a sample of 30000 datapoints from the entire dataset to create a dendrogram. I will not be performing HCA since our systems gave us memory errors. The dendrogram generated shows us either 2,3 or 4 clusters depending on the sample picked.

#### Performing PCA and KMeans

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(data_final)
csum = np.cumsum(pca.explained_variance_ratio_)
np.where(csum>0.95)[0][1]
```

```python
pca = PCA(n_components = 0.95)
X_pca = pca.fit_transform(data_final)
```

```python
from sklearn.cluster import KMeans
wcv = []
for i in range(1,10):
    km = KMeans(n_clusters = i)
    km.fit(X_pca)
    wcv.append(km.inertia_)

plt.figure(figsize = (15,6))
plt.plot(range(1,10),wcv,color = 'g')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
data = pd.DataFrame(X_pca).sample(30000)
plt.figure(figsize = (15,6))
sch.dendrogram(sch.linkage(data,method = 'ward'),p =10,truncate_mode = 'level')
plt.show()
```

```python
km = KMeans(n_clusters = 3)
km.fit(X_pca)
preds = pd.Series(km.predict(X_pca),name = 'Interest_Rate_pred')
```

```python
#To find out which cluster belongs to which class in interest Rate.
```

```python
original_data = pd.read_csv('Data/filled_train_28-03.csv')
new_data = pd.concat((original_data,preds),axis = 1)
new_data.head()
```

```python
#Trying to find out which cluster belongs to which class using all numeric columns
data_num = original_data.select_dtypes(np.number).drop(['Interest_Rate'],axis = 1)
y = original_data['Interest_Rate']
ypred = preds
```

```python
#Boxplots with original Interest_Rate.
fig,ax = plt.subplots(2,4,figsize = (15,6))
for variables,plotsubs in zip(data_num.columns,ax.flatten()):
    plotsubs.set_title(variables)
    sns.boxplot(y,data_num[variables],ax = plotsubs)
    plt.tight_layout()
    plt.subplots_adjust(top = 0.90)
fig.suptitle('Boxplot vs original target variable')
plt.show()
```

```python
#Boxplots with predicted Interest_Rate clusters.
fig,ax = plt.subplots(2,4,figsize = (15,6))
for variables,plotsubs in zip(data_num.columns,ax.flatten()):
    plotsubs.set_title(variables)
    sns.boxplot(ypred,data_num[variables],ax = plotsubs)
    plt.tight_layout()
    plt.subplots_adjust(top = 0.90)
fig.suptitle('Boxplot vs created target variable')
plt.show()
```

###### We see a lot of difference in the characteristics with each of the columns in the boxplots when comparing the boxplots with the predicted target and the boxplots with the original target variable.

```python
y.value_counts()
```

```python
ypred.value_counts()
```

###### I would be labeling the predicted y based on the counts since it seems to be showing a simmilar characteristic.

```python
#0:3
#1:2
#2:1
ypred = ypred.map({0:3,1:2,2:1})
```

```python
compare_tgt = pd.concat((y,ypred),axis = 1)
```

```python
compare_tgt[compare_tgt['Interest_Rate']==compare_tgt['Interest_Rate_pred']]
#We only get 53K matching targets with the original target variable.
```

# Trying on the full train data without removing outliers.

```python
#Performing the same missing value imputations without removing outliers.
d = pd.read_csv('Data/capstone_train.csv')
d = d.drop(['Loan_ID'],axis = 1)

import string
def remove_punctuation(x):
    if len(x.split(','))>1:
        return int(x.split(',')[0]+x.split(',')[1])
    else:
        return int(x)
d['Loan_Amount_Requested'] = d['Loan_Amount_Requested'].apply(remove_punctuation)

d1 = d.copy()
d1 = d1.drop('Months_Since_Deliquency',axis = 1)

#Since Months_Since_Deliquency does not seem to show much of an impact on interest_rates we will drop it.
import time
start = time.time()
#Length_Employed can be filled based on annual_income.
#Annual_Income will be filled using KNNImputer on the numerical columns only.
#Home_Owner will be taken care of later.
knn_impute_data=d1.select_dtypes(np.number)
#Standardizing the dataset for the KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import KNNImputer
ss = StandardScaler()
scaled = pd.DataFrame(ss.fit_transform(knn_impute_data.drop('Interest_Rate',axis = 1)),columns = knn_impute_data.drop('Interest_Rate',axis = 1).columns)
filled_scaled = KNNImputer(n_neighbors = 5).fit_transform(X = scaled,y = knn_impute_data['Interest_Rate'])
end = time.time()
print('Execution Time:',end-start)

filled_unscaled = pd.DataFrame(ss.inverse_transform(filled_scaled),columns = knn_impute_data.drop('Interest_Rate',axis = 1).columns)
fill_data = pd.concat((round(filled_unscaled.reset_index(drop = True)),d1.select_dtypes('object').reset_index(drop = True),d['Months_Since_Deliquency'].reset_index(drop = True),d1['Interest_Rate'].reset_index(drop = True)),axis = 1)

#Filling Length Employed with mode since only 0.04% of values are missing
fill_data['Length_Employed'] = fill_data['Length_Employed'].fillna(fill_data['Length_Employed'].mode().values[0])

#Filling home owner with max type of home for each employment level.
mapper = dict(fill_data.groupby(['Length_Employed','Home_Owner'])['Home_Owner'].count().sort_values().groupby(level = 0).tail(1).index)

import math
for i in fill_data[fill_data['Home_Owner'].isna()].index:
    if math.isnan(fill_data['Home_Owner'][i]) == True:
        fill_data['Home_Owner'][i] = mapper[fill_data['Length_Employed'][i]]

fill_data.to_csv('Data/filled_train_complete_data_28-03.csv',index = False)
```

### Inferences
 * On performing KMeans clustering on the filled train data with outliers removed, we were getting 2-3 as the optimum number of clusters. This tells me that there is a stark difference between 2 categories but there isn't much of a difference in the 2nd and 3rd category.
 
 * After applying PCA and performing predictions we were able to match 37% of the target features. This leads me to believe that this data on its own was not the only factor in predicting the interest rates for many of the applicants.
 
 * Performing the above processes along with the outliers did not make any difference in the overall prediction accuracy which was again approximately 37%.
 
 * We feel that the original target variable seems to require some more features to be able to accurately determine the interest rate for each applicant.
